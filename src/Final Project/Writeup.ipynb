# Cystic Fibrosis and Female Infertility #

## <u>Disease Background<u> ##

> Cystic fibrosis (CF) is a genetic condition resulting from mutations in the cystic fibrosis transmembrane conductance regulator (CFTR) gene.  CFTR encodes a protein that regulates ion transport across epithelial cell membranes. This dysfunction restricts chloride ion transport, causing thick and sticky mucus accumulation in many organ systems, particularly the pulmonary system, the digestive system, the endocrine system, and the reproductive system. The thick mucus is harder to clear from the body than healthy-state mucus, which increases a patients likelihood of developing respiratory infections, experiencing airway inflammation, and developing  lung damage. In the pancreas, CFTR dysfunction obstructs digestive enzyme secretion, leading to malabsorption and nutritional deficiencies.  While CFTR is the primary gene implicated in CF, modifier genes such as TGFB1 and MUC5B are also of interest due to possible influence on disease severity and progression due to their impacts on inflammation and mucus production.

> Most current treatments for CF focus on treating symptoms and decreasing disease severity.  Such treatment includes airway clearance methods, such as chest physiotherapy and bronchodilators. Mucolytic agents are used to thin mucus, while antibiotics target chronic bacterial infections such as Pseudomonas aeruginosa. Malnutrition is often addressed through the use of pancreatic enzyme replacement therapies and nutritional supplements. A recent development for CF treatment is CFTR modulators, such as ivacaftor, lumacaftor, tezacaftor, and elexacaftor, to target specific mutations and restore CFTR function. These modulators have been a game-changer for eligible patients since they target the disease at its root. However, rare CFTR (like nonsense mutations) are not responsive to these modulator therapies, leaving these patients at a disadvantage.  For these patients, the only current strategy is addressing symptoms.

>The incidence rate of CF in the United States is around 1 in 3,500, affecting around 40,000 individuals. The disease is most common among individuals of Northern European descent, and least common in those of Asian descent.  CF occurance is relatively equal between men and women, but disease severity has been reported higher in females, due to biological and hormonal differences. Estrogen levels have been shown to correlate with symptom severity, such as a higher immune response, increasing inflammation, and thicker mucus, further obstructing the airways. Additionally, approximately 35% of women with CF experience significant fertility issues. The primary cause of reduced fertility in women with CF is the production of thick cervical mucus, which can inhibit sperm penetration and reduce the likelihood of fertilization. This is a direct result of CFTR mutations, which disrupt ion and water transport in the reproductive tract, impacting mucus consistency.  The most common way that the fertility issues are addressed for CF women is by utilizing alternative fertilization technologies, such as in-vitro fertilization (IVF) and intrauterine insemination (IUI). 
  
> An improved treatment for CF-related infertility in females would seek to address the production of thick cervical mucus.  In addition to CFTR, other genes that play a role in cervical mucus production would be of interest.  Such genes include:
>> 1.  MUC5B and MUC5AC - responsible for the structural properties of cervical mucus
>> 2.  SCNN1A, SCNN1B, and SCNN1G - encode subunits of epithelial sodium channels, which impact hydration of cervical mucus.
>> 3.  Hormone receptor genes (ESR1 and PGR) - encode estrogen and progesterin receptors that regulate the production and consistency of cervical mucus in response to hormone levels.

> In this study, we seek to investigate the pharmacological network surrounding these genes to identify druggable targets for potential treatment of CF-related female infertility. 

## <u>Methods<u> ##


>A comprehensive literature review identified a set of key genes directly influencing the physical properties of cervical mucus. These genes include CFTR (cystic fibrosis transmembrane conductance regulator, critical for ion transport), MUC5B and MUC5AC (major gel-forming mucins that contribute to the viscoelastic properties of mucus), SCNN1A, SCNN1B, and SCNN1G (subunits of the epithelial sodium channel, ENaC, involved in fluid regulation), as well as ESR1 (estrogen receptor 1) and PGR (progesterone receptor), both of which regulate hormonal responses affecting mucus composition. These genes were input into the STRING database to generate a protein-protein interaction network, as shown in the figure above.

>To identify subclusters within the network, Markov Cluster Algorithm (MCL) clustering was applied with an inflation parameter of 3. This parameter choice allowed for the identification of tighter, more biologically meaningful clusters, as indicated by the dotted lines in the image. The resulting network consisted of 18 interconnected genes, highlighting the complex interplay between hormonal signaling, ion transport, and mucin production in regulating cervical mucus.

>Upon further analysis, eight genes within the network were identified as targets of drugs approved for clinical use. These genes are CFTR, ESR1, PGR, NR3C2 (mineralocorticoid receptor), SCNN1A, SCNN1B, SCNN1G, and SLC12A3 (solute carrier family 12 member 3, also known as the sodium-chloride symporter). Their druggability underscores their therapeutic relevance and potential for clinical interventions.

>In addition, four other genes— SGK1 (serum/glucocorticoid-regulated kinase 1), WNK1 (WNK lysine-deficient protein kinase 1), PRSS8 (prostasin), and KCNJ1 (renal outer medullary potassium channel)—were identified as having high potency but lacking clinically approved drugs. These genes were flagged as promising targets through further analysis using the Pharos database, which evaluates the druggability of genes based on available evidence.

>This combined approach, integrating literature review, network analysis, and database querying, provided a comprehensive overview of potential targets. While the presence of several clinically approved targets is encouraging, the identification of high-potency but currently untapped genes presents opportunities for future drug discovery efforts aimed at modulating cervical mucus properties.

>The search for genes directly contributing to the regulation and composition of cervical mucus was conducted using several online bioinformatics databases, including Pharos, STRING, and HIPPIE. This analysis identified three specific genes of interest: MUC5B (mucin 5B, a major structural component of mucus), SLPI (secretory leukocyte peptidase inhibitor, associated with anti-inflammatory and antimicrobial properties of mucus), and DEFB126 (defensin beta 126, which contributes to mucus's protective barrier and antimicrobial functions).

>To better understand the interactions between these genes and other components of the cervical mucus regulatory network, a secondary interaction map was created in STRING. This map also incorporated CFTR, a previously identified key target, to examine its relationship with MUC5B, SLPI, and DEFB126, as shown in the figure above.

>From this network analysis, CFTR emerged as the only gene linked to an FDA-approved drug, highlighting its clinical relevance. In contrast, all other identified genes, with the exception of DEFB126, were categorized as "Tbio" in the Pharos database. This designation indicates that these genes have been studied to some extent, but no small molecule inhibitors or modulators with established potency have been identified. DEFB126, on the other hand, is classified as "unknown," reflecting a lack of detailed characterization or drug discovery efforts to date.

>Given the limited pool of druggable targets among these genes, direct targeting of cervical mucus-associated genes does not appear to be a promising therapeutic strategy. As a result, the focus of the project will shift to the first network derived from the literature review. This primary network encompasses genes indirectly associated with cervical mucus properties but with a higher likelihood of druggability, providing a more feasible pathway for therapeutic intervention.

## <u>Analysis and Results<u> ##
>The analysis for this project focused on identifying druggable targets relevant to the modulation of cervical mucus properties, as previously discussed. This process began with the identification of key genetic targets—those directly implicated in the regulation of cervical mucus and considered druggable. The primary targets identified included CFTR (cystic fibrosis transmembrane conductance regulator), ESR1 (estrogen receptor 1), PGR (progesterone receptor), SCNN1A, SCNN1G, and SCNN1B (subunits of the epithelial sodium channel, ENaC). These genes are directly involved in pathways regulating the hydration and viscoelastic properties of cervical mucus, making them pivotal to the study's objectives.

>Secondary targets were identified as druggable genes that maintain a direct connection to these primary genes within the analyzed interaction network. This set included NR3C2 (mineralocorticoid receptor), GK1 (glucokinase regulator), WNK1 (WNK lysine deficient protein kinase 1), PRSS8 (prostasin), and KCNJ1 (renal outer medullary potassium channel). These secondary targets represent nodes within the network that, while not directly influencing cervical mucus properties themselves, interact closely with primary targets and thus contribute to the broader regulatory network.

>To assess the viability of these targets, centrality measures—such as degree, betweenness, and closeness centrality—were calculated to prioritize nodes based on their importance within the interaction network. The centrality analysis helped distinguish between key regulatory nodes and those with less significant roles in the pathway.

>Among the targets identified, only WNK1 was found to be druggable using a kinase inhibitor, highlighting its unique potential for therapeutic intervention. This finding emphasizes the limited availability of pharmacological options targeting this network and underscores the importance of WNK1 as a focus for further drug development efforts.


## <u>Discussion<u> ##
>Based on this analysis, a kinase inhibitor targeting WNK1 represents a promising candidate for the treatment of CF-related female infertility. This approach could address infertility through the regulation of sodium channels, which in turn would impact cervical mucus hydration. If a WNK1 kinase inhibitor were to enhance cervical mucus hydration, the mucus would become less viscous, facilitating sperm penetration and increasing fertilization potential.

>Data on hormone levels in CF women would be valuable to further investigate this pathway. Investigating hormone receptors as potential targets shows promise of impacting cervical mucus, given their direct involvement in managing the female reproductive system. However, hormonal treatments are associated with significant side effects, making them less desirable. Further information on the connection between hormones and CF could help determine whether these therapies are feasible or if the adverse effects outweigh any potential benefits.

>The next steps would involve mapping of the complete signaling pathway to evaluate the broader impacts of WNK1 inhibition. While the most direct effect is expected to be increased mucus hydration, it is important to identify any consequences that may impede the viability of this treatment strategy.
